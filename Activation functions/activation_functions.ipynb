## Funkcje Aktywacji

### ReLU Function

$f(x) = max(x, 0))$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
def max_relu(x):
    return max(x , 0)
```

```python
for i in [-10 , -5 , 0, 5 ,10]:
    print(max_relu(i))
```

```python
data = np.arange(-5 , 5 , 0.1)
max_relu_data = [max_relu(x) for x in data]

plt.plot(data , max_relu_data)
plt.title("Relu Function")
```

### Sigmoid Function

$f(x) = \frac{1}{1+e^{-x}}$

```python
def sigmoid(x):
    return 1.0 / (1  + np.exp(-x))
```

```python
data2 = np.arange(-5 , 5 , 0.1)
sigmoid_data = [sigmoid(x) for x in data2]

plt.plot(data2 , sigmoid_data)
plt.title("Sigmoid Function")
```

### Tan h 

$\tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$

```python
def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
```

```python
data3 = np.arange(-5 , 5 , 0.1)
tanh_data = [tanh(x) for x in data3]

plt.plot(data3 , tanh_data)
plt.title("Tan h")
```
